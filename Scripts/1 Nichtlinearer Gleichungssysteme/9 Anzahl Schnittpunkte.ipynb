```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

"""
=======================================================================================================================
INPUT
=======================================================================================================================
"""

# Definiere den Bereich f√ºr x1 und x2
x1_range = np.linspace(-3, 3, 400)
x2_range = np.linspace(-3, 3, 400)

# Erstelle ein Gitter aus x1 und x2 Werten
X1, X2 = np.meshgrid(x1_range, x2_range)

# Definiere die Funktionen F1 und F2
F1 = 1 - X1**2 - X2**2
F2 = ((X1 - 2)**2 / 2) + ((X2 - 1)**2 / 4) - 1

"""
=======================================================================================================================
"""

fig, ax = plt.subplots()

ax.contour(X1, X2, F1, levels=[0], colors='blue')
ax.contour(X1, X2, F2, levels=[0], colors='red')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.grid(True)

plt.show()

def diff_func(x):
    x1, x2 = x
    F1 = 1 - x1**2 - x2**2
    F2 = ((x1 - 2)**2 / 2) + ((x2 - 1)**2 / 4) - 1
    return [F1 - F2, F1 - F2]

x0 = [0, 0]

solution = fsolve(diff_func, x0)

num_intersections = len(solution)

print(f"Die Anzahl der Schnittpunkte ist {num_intersections}.")
```
