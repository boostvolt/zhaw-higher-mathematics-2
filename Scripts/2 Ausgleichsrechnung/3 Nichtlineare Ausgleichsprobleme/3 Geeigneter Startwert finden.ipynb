```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Messdaten
frequencies = np.array([25, 35, 45, 55, 65])
amplitudes = np.array([47, 114, 223, 81, 20])

# Plotten der Datenpunkte
plt.scatter(frequencies, amplitudes, color='blue', label='Messdaten')
plt.xlabel('f / Hz')
plt.ylabel('A(f) / -')
plt.title('Messdaten Resonanzversuch')
plt.legend()
plt.grid(True)
plt.show()

# Bestimmung des Maximums
max_amplitude_index = np.argmax(amplitudes)
f0_start = frequencies[max_amplitude_index]

# Ausgabe des Startwerts
display(Markdown(f"**Geeigneter Startwert f체r den Parameter** $f_0$ **ist:**"))
display(Markdown(f"$f_0 = {f0_start}$ Hz"))
display(Markdown(f"Dieser Startwert wurde gew채hlt, da er der Frequenz entspricht, bei der die Amplitude $A(f)$ maximal ist."))
```
