```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

def mittelpunkt(f, a, b, y0, n=None, h=None):
    if h is None:
        if n is None:
            raise ValueError("Either 'n' or 'h' must be specified.")
        h = (b - a) / n
    else:
        n = int((b - a) / h)

    h2 = h / 2
    x = a
    y = y0

    X = np.zeros(n + 1)
    Y = np.zeros(n + 1)

    X[0] = x
    Y[0] = y

    for i in range(0, n + 1):
        x2 = x + h2
        y2 = y + h2 * f(x, y)
        x_next = x + h
        y_next = y + h * f(x2, y2)
        
        display(Math("\\text{{Schritt {}}}".format(i + 1)))
                
        # Berechnung von x_{h/2}
        display(Math("x_{{\\frac{{{}}}{{2}}}} = x_{} + \\frac{{h}}{{2}} = {} + \\frac{{{}}}{{2}} = {}".format(h, i, x, h, x2)))

        # Berechnung von y_{h/2}
        display(Math("y_{{\\frac{{{}}}{{2}}}} = y_{} + \\frac{{h}}{{2}} \\cdot f(x_{}, y_{}) = {} + \\frac{{{}}}{{2}} \\cdot f({}, {}) = {}".format(h, i, i, i, y, h, x, y, y2))) 

        # Berechnung von x_{i+1}
        display(Math("x_{} = x_{} + h = {} + {} = {}".format(i+1, i, x, h, x_next)))

        # Berechnung von y_{i+1}
        display(Math("y_{} = y_{} + h \\cdot f(x_{{\\frac{{{}}}{{2}}}}, y_{{\\frac{{{}}}{{2}}}}) = {} + {} \\cdot f({}, {}) = {}".format(i+1, i, h, h, y, h, x2, y2, y_next)))

        x = x_next
        y = y_next
        X[i] = x
        Y[i] = y

        print()

    print("All x values:", ', '.join(map(str, X.flatten().tolist())))
    print("All y values:", ', '.join(map(str, Y.flatten().tolist())))

    return X, Y

"""
=======================================================================================================================
INPUT
=======================================================================================================================
"""

def f(x, y):
    return (x**2 / y)

a = 0
b = 1.4
y0 = 2

# Entweder Schrittanzahl n oder Schrittweite h angeben
X, Y = mittelpunkt(f, a, b, y0, n=2)
# X, Y = mittelpunkt(f, a, b, y0, h=0.2)

"""
=======================================================================================================================
"""

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(X, Y, label='Mittelpunkt method')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of differential equation using Mittelpunkt method')
plt.legend()
plt.grid(True)
plt.show()
```
