```python
import numpy as np
import matplotlib.pyplot as plt

#Intervall
a = 1.
b = 4.

#Schrittweite und Anzahl Schritte
h = 0.02
n = int((b-a)/h)

#Anzahl Zeilen im Vektor
rows = 4

x = np.zeros(n+1)
z = np.zeros([rows,n+1])
k = np.zeros([rows,n+1])

x[0] = a

#Startvektor
z[:,0] =np.array([0.,3.,5.,8.])
k[:,0] =np.array([0.,3.,5.,8.])

#Umgewandelte höhere Ordnung in 1. Ordnung
def f(x,z): 
    return np.array(
        [
            z[1], 
            z[2], 
            z[3], 
            6*x + 3*z[3]/x - 6*z[2]/x**2 + 6*z[1]/x**3
        ])

for i in range(0,n):
    xh2 = x[i] + h / 2
    yh2 = z[:,i] + h / 2 * f(x[i], z[:,i])

    x[i+1]=x[i]+h
    z[:,i+1]=z[:,i]+h*f(xh2, yh2)

h = 0.02

for i in range(0,n):
    xh2 = x[i] + h / 2
    yh2 = k[:,i] + h / 2 * f(x[i], k[:,i])

    x[i+1]=x[i]+h
    k[:,i+1]=k[:,i]+h*f(xh2, yh2)

def exact(x):
    return 2 * x + x ** 2 - x ** 3 + x ** 4

plt.plot(x, exact(x), 'o')
plt.plot(x,z[1,:],x,z[2,:],x,z[3,:]), plt.legend(["Lösung y(x)", "y'(x)","y''(x)","y'''(x)"])   
  
```
